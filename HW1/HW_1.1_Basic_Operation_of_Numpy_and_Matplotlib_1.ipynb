## Question 1
Print the result of **A+B** and explain the mechanism behind this. 

You answer: 

```python
A = np.array([[0, 1, 2]])
B = np.array([[10, 10, 10], [20, 20, 20], [30, 30, 30]])
# insert your code and provide a description of the underlying mechanism called broadcast
print(A+B)
#Broadcasting helps match the sizes of two differently sized arrays.
#The smaller array expands until it reaches the dimensions of the smaller array.
#In this example, array A is expanded into [[0, 1, 2], [0, 1, 2], [0, 1, 2]].
#This expanded array is then added to B to get the result.
```

## Question 3 
Generate an array with 100 rows and 2 cols at random. Plot it by Matplotlib. You should consider the plotting setting in terms of titles, legends, names of the x and y axe, etc.

**Firstly, please set your student id as the seed number. Like np.random.seed(20016329).** 

```python
np.random.seed(20011191) # Please use your student id as the seed number
data = np.random.rand(100, 2) #Creates an array with 100 rows and 2 columns
plt.scatter(data[:, 0], data[:, 1], color='black', edgecolors='black') #Generates the scatter plot
plt.title("Scatter Plot of Randomly Generated Data from a uniform distribution over [0, 1]") #Creates ta title for the scatter plo
plt.xlabel("Col 1") #Labels the x-axis
plt.ylabel("Col 2") #Labels the y-axis
plt.show() #Shows the plot
```

## Question 4
For the previous generated data, consider each row of the array as an individual data point, with the two columns representing its distinct features. Identify the data points with the smallest and largest sum of features. Mark them in red and blue in the figure respectively.

Hint: You can compute the sum of each row and store it in an array named "sums". Then find the minimum or maximum index of this array. 

```python
sums = np.sum(data, axis=1); #Sums each row of the array
min = np.argmin(sums) #Finds the minimum sum of the array
max = np.argmax(sums) #Finds the maximum sum of the array

plt.scatter(data[:, 0], data[:, 1], color = "black", label = "Data Points") #Generates the plot 
plt.scatter(data[min, 0], data[min, 1], color = "red", label = "Smallest Sum") #Genereates the lowest sum in a blue color
plt.scatter(data[max, 0], data[max, 1], color = "blue", label = "Largest Sum") #Generates the largest sum in a red color

plt.title("Data Points with Extremes Marked") #Creates a title for the plot
plt.xlabel("Feature 1") #Labels the x-axis
plt.ylabel("Feature 2") #Labels the y-axis
plt.legend(loc="upper center", bbox_to_anchor=(0.5, -0.11), ncol=3) #Generates the legend for the plot

plt.show() #Shows the plot
```

## Question 5
Add two lines representing the mean of the Feature 1 and the 25% quantile of the Feature 2 seperately. Add the legend of the two lines.

```python
mean = np.mean(data[:, 0]) #Finds the mean of the x-axis
quant = np.quantile(data[:, 1], 0.25) #Finds the 25% quantile of the y-axis

plt.scatter(data[:, 0], data[:, 1], color = "black", label = "Data Points") #Generates the plot
plt.axvline(mean, color = "green", label = "Mean of Col 1") #Creates the horizontal line for the mean of the x-axis
plt.axhline(quant, color = "orange", label = "25% Quantile of Col 2") #Creates the vertical line for the mean of the y-axis

plt.title("Scatter Plot with Mean and Quantile Plots") #Generates the title for the plot
plt.xlabel("Col 1") #Labels the x-axis
plt.ylabel("Col 2") #Labels the y-axis
plt.legend(loc="upper center", bbox_to_anchor=(0.5, -0.11), ncol=3) #Generates the legend for the plot
plt.show() #Shows the plot
```
